```python
import numpy as np
import pandas as pd
```

```python
df = df = pd.read_csv("C:\\Users\\91636\\OneDrive\\Desktop\\Regex ML\\Data\\click.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['Ad Topic Line'] = lb.fit_transform(df['Ad Topic Line'])
df['City'] = lb.fit_transform(df['City'])
df['Gender'] = lb.fit_transform(df['Gender'])
df['Country'] = lb.fit_transform(df['Country'])
df['Timestamp'] = lb.fit_transform(df['Timestamp'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func = np.log1p)
```

```python
new_df = ft.fit_transform(df)
```
